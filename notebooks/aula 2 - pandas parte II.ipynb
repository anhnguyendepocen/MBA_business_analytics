## *Estatísticas para analise univariada:*

### Para uma variável $X$ com $n$ amostras $x_1, x_2,..., x_n$

### Média

$$\overline{x}_n = \frac{1}{n} \sum_{i=1}^n x_i $$

### Desvio padrão ($\sigma$) e variancia ($\sigma^2$)

$$\sigma = \sqrt{ \frac{1}{n-1} \sum_{i=1}^n (x_i - \overline{x}_n)^2 }$$

### Mediana

valor que valor numérico que separa a metade superior de uma amostra de dados ordenada, da metade inferior

## *Estatísticas para analise bivariada:*

### Covariancia

$$cov_{x,y}=\frac{\sum_{i=1}^{N}(x_{i}-\bar{x})(y_{i}-\bar{y})}{n-1}$$

### Correlação (Pearson's r)

$$r = \frac{cov_{x,y}}{\sigma_x \sigma_y} = \frac{\sum_{i=1}^{N}(x_{i}-\bar{x})(y_{i}-\bar{y})}
{ \sqrt{ \sum_{i=1}^n (x_i - \overline{x}_n)^2 }
  \sqrt{ \sum_{i=1}^n (y_i - \overline{y}_n)^2 } }$$

## agrupamentos com groupby

essa funcionalidade, bastante útil, sendo largamente empregada na criação de features. Essas features são funções de valores únicos de colunas, as colunas escopo do groupby.

vejamos inicialmente a mecanica como pandas opera os grupos encontrados.<br>
Vamos supor que desejamos estabelecer alguma operação sobre equipes pertencentes a mesma UF.

o método `.groupby()` aplicado sobre a coluna `UF` gera um objeto iterável, que vamos chamar de `df_agrupado`.

```python
df_agrupado = df.groupby('UF')
```

veja que repartimos o dataframe em fragmentos, onde cada fragmento possui um valor único para `UF`.

```python
for key, parte_do_df in df_agrupado:
    print(key)
    print(parte_do_df)
    print('-' * 30)
```

```python
type(df_agrupado)
```

Sobre um objeto agrupado (tipo `DataFrameGroupBy`), podemos chamar métodos que atuam sobre as colunas, mas cuja operação se restringe aos fragmentos do dataframe original, ou seja sobre os grupos. <br>
Por exemplo, se quisermos saber a média de pontos por `UF`, utilizariamos o método `.mean()`:

```python
df_agrupado.PTS.mean()#.sort_values(ascending = False)
```

notar que essa operação faz com que os elementos únicos de `UF` se tornem indices de um novo objeto: uma pandas `Series`.

Similarmente, `std()` fornece o desvio padrão.

```python
df_agrupado.PTS.std()
```

#### agg()
a função `.agg()` permite passar como argumentos, multiplas funções a serem executadas sobre colunas. Essa função admite algumas opções como:

* uma única função, identificada por um ponteiro para função ou alguns strings pré-determinados:

```python
df_agrupado.PTS.agg('mean')
```

* uma lista com ponteiros para funções ou strings

```python
df_agrupado.PTS.agg(['mean', 'std', np.median])
```

* a função pode ser definida e chamada pelo seu nome

```python
def max_min(arr):
    return arr.max() - arr.min()
```

```python
df_agrupado.PTS.agg(max_min)
```

```python
df_agrupado.PTS.agg(['mean', max_min])
```

* finalmente, o argumento pode ser um dicionário onde:  
    1. chave é o nome da coluna; 
    2. valor é a função a ser chamada sobre as colunas

```python
df_agrupado.agg({'PTS': ['mean', 'median'], 'V': ['mean', 'median']})
```

notar que serão gerados colunas com indices hieraquicos.

#### apply()
`.apply()` opera sobre todo o dataframe, segundo seus eixos. `axis = 0` é a opção default. <br>
Ou seja, a função que vai no argumento é aplicada sobre os valores das colunas(\*), todas elas.
Na maioria das vezes, seu resultado é similar ao `.agg()`, quando designamos apenas uma função.

(\*) assim como já explicado, a iteração sobre elementos de `axis = 0` (linhas), significa somar os valores das colunas. <br>
Similarmente, a iteração sobre elementos de `axis = 1` (colunas), significa somar os valores das linhas.

```python
df_agrupado.apply(np.mean)
```

apesar dos resultados não trazerem nenhum sentido, podemos também chamar `df_agrupado.apply(np.mean, axis = 1)`. <br>
Nesse caso, os valores médios serão tomados por cada linha. 

#### transform()
utilizamos `.transform()` quando :

* precisamos incluir uma nova coluna cujo valor depende de um agrupamento por valor
* e portanto o objeto resultante tem com as mesmas dimensões do dataframe original (o mesmo shape).

por exemplo, se quisermos saber o ranking da equipe em sua UF, podemos criar uma nova coluna:

```python
df['ranking_na_UF'] = df_agrupado.PTS.transform(lambda x: range(1, len(x) + 1)).astype(int)
```

vamos aprender como juntar dois dataframes. De maneira generica existem:

### merge

```python
df.index.name = 'Equipe'
df.reset_index(inplace = True)
```

```python
df_valor.head(3)
```

vamos fazer um `merge` primeiro. veja os parametros escolhidos, que correspondem a como os dois dataframes serão juntados:

* `how = 'left'`
* `right_on = 'Clube'`
* `left_on = 'Equipe'`

Mas repare uma coisa, o dataframe resultante tem 22 linhas!! O que aconteceu?

```python
pd.merge(df, df_valor, how = 'left', right_on= 'Clube', left_on = 'Equipe' )
```

sim, temos uma chave duplicada: `Atlético`. Como pandas não tem como saber qual dos dois `Atlético`'s juntar, acaba por juntar ambos duplicando os valores do dataframe da esquerda. 

Se a parametrização escolhida tivesse sido `how = 'right'`, o resultado seria o mesmo, mas com outra ordem de linhas.

para remediar esse problema, vamos renomear as equipes:

```python
df.loc[1 , 'Equipe'] = 'Atlético MG'
df.loc[18 , 'Equipe'] = 'Atlético PR'

df_valor.loc[9, 'Clube'] = 'Atlético MG'
df_valor.loc[11, 'Clube'] = 'Atlético PR'
```

```python
df_tudo = pd.merge(df, df_valor, how = 'left', right_on= 'Clube', left_on = 'Equipe' )
```

* existe correlação entre valor de mercado total e pontos ganhos?
* ou entre derrotas e idade?

```python
sns.heatmap(df_tudo.corr(), cmap = 'coolwarm')
```
